```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('realty_data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df[['price', 'rooms','floor', 'lat', 'lon']]
df.dropna(inplace=True)
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df.drop(columns='price'),
    df['price'],
    random_state=2025,
    test_size=0.3
)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
from sklearn.metrics import r2_score, mean_absolute_percentage_error
lr.fit(X_train, y_train)
y_preds_lr = lr.predict(X_test)

print(f'MAPE: {mean_absolute_percentage_error(y_test, y_preds_lr)}')
print(f'R2-score: {r2_score(y_test, y_preds_lr)}')
```

```python
data = {
    'rooms': [1],
    'floor' : [1],
    'lat': [55.5],
    'lon': [37],
}

probe = pd.DataFrame(data)
pred = lr.predict(probe)[0]
print(f'Approximate price of real estate: {int(pred)} rubles.')
```
