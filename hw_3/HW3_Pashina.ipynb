```python
survival_by_class = df.groupby('Pclass')["Survived"].sum()/df['Pclass'].value_counts()*100
for i in survival_by_class.index:
    print(f'In the {i} class the survival rate of passengers is {round(survival_by_class.values[i-1],2)} %.')
```

```python
#df['Only_name'] = 

df['Only_name'] = df['Name'].str.split(' ').str[2]
df2 = df[df['Sex']=='male']['Only_name'].value_counts()
print(f'The most popular male name is {str(list(df2[0:1].index)[0]).strip()}. {str(df2[0:1].values[0])} men have this name!')
```

```python
df.loc[df['Name'].str.find('Mrs.')!=-1,'Only_name_female'] = df.loc[df['Name'].str.find('Mrs.')!=-1,'Name'].str.split('(').str[1].str.split(' ').str[0]
df.loc[df['Name'].str.find('Mrs.')==-1,'Only_name_female'] = df.loc[df['Name'].str.find('Mrs.')==-1,'Name'].str.split(' ').str[2]
df2 = df[df['Sex']=='female'].groupby('Pclass')['Only_name_female'].value_counts()
for i in survival_by_class.index:
    print(f'In the {i} class the most popular female name is {str(list(df2.loc[i][0:1].index)[0]).strip()}. { str(df2.loc[i][0:1].values[0])} women have this name!')
```
